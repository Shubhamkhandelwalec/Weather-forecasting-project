```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
data=pd.read_csv('daily_weather.csv')
```

```python
data.columns
```

```python
data
```

```python
data.head()
```

```python
data=data.dropna()
data
```

```python
clean_data=data
clean_data['high_humidity_label']=(clean_data['relative_humidity_3pm']>24.99)*1
clean_data.head()
```

```python
X=clean_data.drop(['relative_humidity_9am','relative_humidity_3pm','high_humidity_label','number'],axis=1)
X.head()
```

```python
Y=clean_data['high_humidity_label']
Y.head()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33)
```

```python
humidity_clf=DecisionTreeClassifier(max_leaf_nodes=10,random_state=2)
humidity_clf.fit(X_train,Y_train)
```

```python
pred_test=humidity_clf.predict(X_test)
```

```python
from pandas import DataFrame
```

```python
import numpy as np
```

```python
pd.concat([DataFrame(np.ravel(Y_test),columns=['Test_data']),DataFrame(pred_test,columns=['predicted_test'])],axis=1)
```

```python
accuracy_score(Y_test,pred_test)
```

```python
pred_train=humidity_clf.predict(X_train)
pd.concat([DataFrame(np.ravel(Y_train),columns=['Test_data']),DataFrame(pred_train,columns=['predicted_test'])],axis=1)

```

```python
accuracy_score(Y_train,pred_train)
```

```python
humidity_clf=LogisticRegression()
humidity_clf.fit(X_train,Y_train)
```

```python
pred_test=humidity_clf.predict(X_test)
pd.concat([DataFrame(np.ravel(Y_test),columns=['Test_data']),DataFrame(pred_test,columns=['predicted_test'])],axis=1)
```

```python
accuracy_score(Y_test,pred_test)
```
